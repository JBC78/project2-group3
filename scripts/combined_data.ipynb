```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
```

```python
population = pd.read_csv("../Resources/Population Estimates by County.csv")
population.drop(population.tail(1).index,inplace=True)
population.drop(population.head(1).index,inplace=True)
# population = population[1:len(population)-1]
population.rename(columns={'Population Estimates by County':'County', 'Unnamed: 1': 'Population'}, inplace=True)

population['County'] = population['County'].str.replace('County', '')
population['Population'] = population['Population'].str.replace(',', '')
# population['Population'] = population['Population'].astype(int)
population.reset_index(inplace=True)
population.drop(columns = ['index'], inplace = True)
population.info()
```

```python
parks = pd.read_csv("../Resources/state-park-facility-points-1.csv")
parks.drop(columns=['Category', 'Region', 'Golf','Camp', 'Playground', 'Facility URL',\
     'Nature Center', 'Beach', 'Golf URL', 'Nature Center URL', 'Location',\
    'Longitude', 'Latitude'], inplace = True)
parks_grouped = parks.groupby('County').count()
parks_grouped.reset_index(inplace=True)
```

```python
all_attendance = pd.read_csv("../Resources/State_Park_Annual_Attendance_Figures_modified.csv")
sum_attendance_grouped = all_attendance.groupby(['Facility', 'County']).sum().reset_index()
sum_attendance_grouped.drop(columns=['Year'], inplace = True)
sum_attendance_grouped[sum_attendance_grouped['County'] == 'Bronx']
```

```python
count_attendance_grouped = sum_attendance_grouped.groupby(['County']).count().reset_index()
count_attendance_grouped.drop(columns=['Attendance'], inplace = True)
count_attendance_grouped['County'] = count_attendance_grouped['County'].astype(str)
count_attendance_grouped.info()
```

```python
# Join population and attendance
park_pop_fac = pd.merge(
    count_attendance_grouped,
    population,
    on='County',
    how="right"
)
park_pop_fac
```
